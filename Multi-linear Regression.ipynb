# Import Libaries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Load Data

```python
df=pd.read_csv('C:\\Users\\Nouman Ahmad\\Desktop\\Machine Learning\\Regression\\datasets\\weight-height.csv')
df.head(5)
```

# Data Exploration For Males

```python
male_df=df[df['Gender']=='Male']
male_x=male_df['Height']
male_y=male_df['Weight']
plt.scatter(male_x,male_y)
plt.title('Relationship between Males Height and Weight')
# plt.xlable('Height(inches)')
# plt.ylable('Weight(pounds)')
plt.show()
```

# Data Exploration For Females

```python
male_df=df[df['Gender']=='Female']
male_x=male_df['Height']
male_y=male_df['Weight']
plt.scatter(male_x,male_y,color='red')
plt.title('Relationship between Fmales Height and Weight')
# plt.xlable('Height(inches)')
# plt.ylable('Weight(pounds)')
plt.show()
```

# Now its time to Preprocessing the data

```python
df_dummy=pd.get_dummies(df)
```

```python
df_dummy.head()
```

```python
df_dummy.drop(['Gender_Female'],axis=1, inplace=True)
df_dummy.rename(columns={'Gender_Male':'Gender'},inplace=True)
df_dummy.head()
```

# Now make a model

```python
mlr=LinearRegression()
```

```python
x=df_dummy[['Height','Gender']]
y=df_dummy['Weight']
mlr.fit(x,y)
```

```python
print(mlr.intercept_)
print(mlr.coef_)
```
